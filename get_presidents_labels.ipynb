```python
import sys
```

```python
mypath = './repos/President-Speeches-Embeddings/president_scraper/data/'
```

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
onlyfiles
```

```python
files = []
```

```python
for direc in listdir(mypath):
    files.append(direc)
```

```python
pp_dict = {}
```

```python
for y in files:
    for x in df.idno_name.values:
        if y in x.lower().split("_")[1]:
            pp_dict[y] = x
            break
        else:
            pp_dict[y] = 'None'
```

```python
# George Washington
# John Adams
# William Harrison
# Zachary Taylor
# Millard Fillmore
# Andrew Johnson
# Rutherford Hayes
# James Garfield
# Herbert Hoover
```

```python
pp_dict['bush'] = '99908_ BUSH         '
pp_dict['gwbush'] = '99910_ BUSH         '
pp_dict['lbjohnson'] = '99903_ JOHNSON      '
pp_dict['jqadams'] = '99874_ ADAMS J Q    '
pp_dict['bharrison'] = '99891_ HARRISON B   '
pp_dict['vanburen'] = '99876_ VAN BUREN    '
pp_dict['fdroosevelt'] = '99899_ ROOSEVELT F  '

pp_dict['washington'] = 'None'
pp_dict['adams'] = 'None'
pp_dict['harrison'] = 'None'
pp_dict['taylor'] = 'None'
pp_dict['fillmore'] = 'None'
pp_dict['johnson'] = 'None'
pp_dict['hayes'] = 'None'
pp_dict['garfield'] = 'None'
pp_dict['hoover'] = 'None'
```

```python
df.idno_name.values
```

```python
pp_dict
```

```python
with open('folders_to_presidents.csv','w') as f:
    f.write("foldername,idno_name\n")
    for x in pp_dict.items():
        f.write(x[0]+","+x[1]+"\n")
```

### Regression to classification (segmenting labels)

```python
!head -n 10 presidents_with_labels.csv
```

```python
df = pd.read_csv('presidents_with_labels.csv', index_col=0)
```

```python
df.head()
```

```python
def get_label(num):
    if num >= -1. and num < -0.75:
        return 0 #ultra-liberal
    elif num >= -.75 and num < -0.5:
        return 1 #very-liberal
    elif num >= -.5 and num < -0.25:
        return 2 #liberal
    elif num >= -.25 and num < 0.:
        return 3 #somewhat-liberal
    elif num >= 0. and num < 0.25:
        return 4 #somewhat conservative
    elif num >= .25 and num < 0.5:
        return 5 #conservative
    elif num >= .5 and num < 0.75:
        return 6 #very-conservative
    elif num >= .75 and num <= 1.:
        return 7 #ultra-conservative
```

```python
df.dwnom1
```

```python
df['label'] = df.dwnom1.apply(lambda x: get_label(x)) #get_label(df['dwnom1'])
```

```python
df.head()
```

```python
df.to_csv('presidents_with_labels.csv')
```
